```python
# 2. Entropy & Length
from sklearn.utils import shuffle
import pandas as pd

data = pd.read_csv('featured.csv', delimiter=',')
df = shuffle(data)
```

```python
df.head()
```

```python
df_malicious = df.loc[df['is_malicious'] == 1]  # malicious
df_genuine = df.loc[df['is_malicious'] == 0]  # genuine
```

```python
df_malicious.head()
```

```python
df_genuine.head()
```

```python
feat_Table_1 = df_malicious.head(25000).append(df_genuine.head(25000))   # malicious
feat_Table_1.head(10)
```

```python
feat_Table_2 = df_genuine.head(25000).append(df_malicious.head(25000))  # genuine
feat_Table_2.head(10)
```

```python
# Set figsize here
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

pal=sns.diverging_palette(240, 10, n=2, center="light")
#flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
#sns.palplot(sns.color_palette(flatui))

# fig.set_size_inches(12, 5)
sns.scatterplot(x="url_entropy", y="url_length", hue='is_malicious', data=feat_Table_1, ax=axes[0][0], palette=pal) # malicious, genuine
sns.scatterplot(x="url_entropy", y="url_length", hue='is_malicious', data=feat_Table_2, ax=axes[0][1], palette=pal) # genuine, malicious
sns.scatterplot(x="path_entropy", y="path_length", hue='is_malicious', data=feat_Table_1, ax=axes[1][0], palette=pal)
sns.scatterplot(x="path_entropy", y="path_length", hue='is_malicious', data=feat_Table_2, ax=axes[1][1], palette=pal)
```

```python
# 3. N-gram
feat_Table_31 = df_malicious.head(25000).append(df_genuine.head(25000))  # malicous, genuine
feat_Table_32 = df_genuine.head(25000).append(df_malicious.head(25000))  # genuine, malicous
```

```python
# Set figsize here malicious
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(24,16))

pal=sns.diverging_palette(240, 10, n=2, center="light")

# fig.set_size_inches(12, 5)
sns.scatterplot(x="unigram_avg", y="unigram_std", hue='is_malicious', data=feat_Table_31, ax=axes[0][0], palette=pal)
sns.scatterplot(x="bigram_avg", y="bigram_std", hue='is_malicious', data=feat_Table_31, ax=axes[0][1], palette=pal)
sns.scatterplot(x="trigram_avg", y="trigram_std", hue='is_malicious', data=feat_Table_31, ax=axes[0][2], palette=pal)

sns.scatterplot(x="unigram_avg", y="unigram_std", hue='is_malicious', data=feat_Table_32, ax=axes[1][0], palette=pal)
sns.scatterplot(x="bigram_avg", y="bigram_std", hue='is_malicious', data=feat_Table_32, ax=axes[1][1], palette=pal)
sns.scatterplot(x="trigram_avg", y="trigram_std", hue='is_malicious', data=feat_Table_32, ax=axes[1][2], palette=pal)
```

```python
# 5. Linguistic 
# Read file - Dataframe
feat_Table_l = pd.read_csv('feature12.txt',delimiter='\t') 
feat_Table_l.head(5)
```

```python
feat_Table_l_m = feat_Table_l.tail(25000).append(feat_Table_l.head(25000))   # malicious
feat_Table_l_m.head(5)
```

```python
feat_Table_l_g = feat_Table_l.head(25000).append(feat_Table_l.tail(25000))   # genuine
feat_Table_l_g.head(5)
```

```python
# Set figsize here
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(24,10))
pal=sns.diverging_palette(240, 10, n=2, center="light")

# fig.set_size_inches(12, 5)
sns.scatterplot(x="vowel_ratio", y="length", hue='is_malicious', data=feat_Table_l_m, palette=pal, ax=axes[0][0])
sns.scatterplot(x="digit_ratio", y="length", hue='is_malicious', data=feat_Table_l_m, palette=pal, ax=axes[0][1])
sns.scatterplot(x="repeat_letter", y="length", hue='is_malicious', data=feat_Table_l_m, palette=pal, ax=axes[0][2])
sns.scatterplot(x="consec_digit", y="length", hue='is_malicious', data=feat_Table_l_m, palette=pal, ax=axes[0][3])
sns.scatterplot(x="consec_consonant", y="length", hue='is_malicious', data=feat_Table_l_m, palette=pal, ax=axes[0][4])

sns.scatterplot(x="vowel_ratio", y="length", hue='is_malicious', data=feat_Table_l_g, palette=pal, ax=axes[1][0])
sns.scatterplot(x="digit_ratio", y="length", hue='is_malicious', data=feat_Table_l_g, palette=pal, ax=axes[1][1])
sns.scatterplot(x="repeat_letter", y="length", hue='is_malicious', data=feat_Table_l_g, palette=pal, ax=axes[1][2])
sns.scatterplot(x="consec_digit", y="length", hue='is_malicious', data=feat_Table_l_g, palette=pal, ax=axes[1][3])
sns.scatterplot(x="consec_consonant", y="length", hue='is_malicious', data=feat_Table_l_g, palette=pal, ax=axes[1][4])
```
